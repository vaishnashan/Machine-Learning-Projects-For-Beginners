```python
data=data.drop(["school","age"], axis=1)
```

```python
data.head()
```

```python
data.head()
```

```python
data.shape
```

```python
#Converting to categorical value
categorical_d = {'yes': 1, 'no': 0}
data['schoolsup'] = data['schoolsup'].map(categorical_d)
data['famsup'] = data['famsup'].map(categorical_d)
data['paid'] = data['paid'].map(categorical_d)
data['activities'] = data['activities'].map(categorical_d)
data['nursery'] = data['nursery'].map(categorical_d)
data['higher'] = data['higher'].map(categorical_d)
data['internet'] = data['internet'].map(categorical_d)
data['romantic'] = data['romantic'].map(categorical_d)
```

```python
categorical_d = {'F': 1, 'M': 0}
data['sex'] = data['sex'].map(categorical_d)

# map the address data
categorical_d = {'U': 1, 'R': 0}
data['address'] = data['address'].map(categorical_d)

# map the famili size data
categorical_d = {'LE3': 1, 'GT3': 0}
data['famsize'] = data['famsize'].map(categorical_d)

# map the parent's status
categorical_d= {'T': 1, 'A': 0}
data['Pstatus'] = data['Pstatus'].map(categorical_d)

# map the parent's job
categorical_d = {'teacher': 0, 'health': 1, 'services': 2,'at_home': 3,'other': 4}
data['Mjob'] = data['Mjob'].map(categorical_d)
data['Fjob'] = data['Fjob'].map(categorical_d)

# map the reason data
categorical_d= {'home': 0, 'reputation': 1, 'course': 2,'other': 3}
data['reason'] = data['reason'].map(categorical_d)

# map the guardian data
categorical_d = {'mother': 0, 'father': 1, 'other': 2}
data['guardian'] = data['guardian'].map(categorical_d)
```

```python
data.head(90)
```

```python
from sklearn.model_selection import train_test_split
x=data.drop("G3",axis=1)
y=data['G3']
```

```python
data["G3"]
```

```python

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.20, random_state=44)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
L_Model=LinearRegression()
```

```python

L_Model.fit(X_train, y_train)
```

```python
y_pred=L_Model.predict(X_test)
```

```python
y_pred
```

```python
print(L_Model.score(X_test, y_test)) 
```
