***1. Line Plot***

***A line plot shows the relationship between two variables, with a line connecting the data points.***

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.plot(Data['YearsExperience'], Data['Salary'], marker='o', linestyle='-')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary')
plt.grid(True)
plt.show()
```

***2. Scatter Plot***

***A scatter plot shows individual data points and is useful for seeing the distribution and relationship between two variables.***

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(Data['YearsExperience'], Data['Salary'], c='blue', marker='o')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = Data.drop('Salary',axis = 1)
y = Data['Salary']
```

```python
x
```

```python
y
```

```python
x.head()
```

```python
y.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42) 
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
L_Model=LinearRegression()
```

```python
L_Model.fit(xtrain,ytrain)
```

```python
prediction=L_Model.predict(xtest)
prediction
```

```python
print(L_Model.score(xtest, ytest)) 
```
