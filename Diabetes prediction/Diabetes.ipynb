TRAIN AND TEST


```python
import sklearn
from sklearn.model_selection import train_test_split  # for this train and test split function we need to parameters , they are data and label , or it means input and output
```

```python
X=Data.iloc[ : ,Data.columns!="Outcome"] #df.iloc[:, df.columns != "Outcome"] selects all rows and columns where the column name is not "Outcome".
Y=Data.iloc[ : ,Data.columns=="Outcome"]  # df.iloc[:, df.columns == "Outcome"] selects all rows and the column where the column name is "Outcome".
```

```python
print(X)
```

```python
print(Y)
```

A train-test split function typically requires two main variables: the features (data) and the labels (target variable). However, additional parameters are often used to control the splitting process.


Basic Usage

Required Variables

Features (X): The input data (independent variables) used to train the model.

Labels (Y): The target variable (dependent variable) used to evaluate the model.

Additional Parameters


test_size: Proportion of the dataset to include in the test split (e.g., 0.2 means 20% of the data will be used for testing).

train_size: Proportion of the dataset to include in the train split (optional if test_size is specified).

random_state: Controls the shuffling applied to the data before splitting (useful for reproducibility).


shuffle: Whether or not to shuffle the data before splitting (default is True).



Output

The function returns four variables:

X_train: Training data for features.

X_test: Testing data for features.

Y_train: Training data for labels.

Y_test: Testing data for labels.


```python
xtrain,xtest,ytrain , ytest =train_test_split(X,Y,test_size=0.2)  #20% of the data for test
```

RANDOMFOREST ALGORITHM


```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Model_RF=RandomForestClassifier()
```

```python
Model_RF.fit(xtrain,ytrain) # to train the algorithm #model.fit(xtrain,ytrain.values.ravel())
```

```python
Model_RF.predict(xtest) # to test the algorithm using test data
```

```python
prediction=Model_RF.predict(xtest)
```

```python
prediction
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(prediction,ytest)

```
